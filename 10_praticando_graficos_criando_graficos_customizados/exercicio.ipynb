#### Exercício 1: Criando o primeiro gráfico

Vamos começar nossa jornada na criação e customização de gráficos com esta atividade, que é apenas o primeiro passo. Aqui, você terá a oportunidade de analisar um conjunto de dados sobre aluguéis de apartamentos em diferentes regiões de São Paulo, explorando como a região influencia o valor médio do aluguel. Essa análise ajudará a identificar padrões ou tendências entre as regiões e os preços de aluguel, fornecendo insights valiosos sobre o mercado imobiliário da cidade.

Sua tarefa será:
- Carregar os dados para análise;
- Agrupar os dados por região e calcular a média dos valores de aluguel;
- Ordenar as regiões em ordem decrescente, do maior para o menor valor médio de aluguel;
- Criar uma figura e definir um eixo, ajustando o tamanho da figura para que os dados fiquem bem visualizados;
- Escolher um gráfico apropriado para comparar os valores médios de aluguel entre as regiões;
- Adicionar um título ao gráfico para torná-lo mais informativo e fácil de interpretar.

Dica: Ao adicionar o título, considere:
- Aumentar o tamanho da fonte para dar destaque;
- Inserir um espaçamento adequado entre o título e o gráfico para melhorar a leitura;
- Posicionar o título no lado esquerdo, pois isso segue o fluxo natural da leitura em idiomas que utilizam o sentido da esquerda para a direita.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'],y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

plt.show()
```

#### Exercício 2: Removendo a saturação do gráfico

Agora é o momento de aprimorar nossa visualização, focando em torná-la mais clara e eficiente. Para isso, vamos trabalhar na remoção da saturação visual, eliminando elementos que não agregam valor ou que podem distrair a atenção.

Sua tarefa será:
- Ocultar o rótulo do eixo X para um visual mais limpo;
- Aumentar o tamanho das fontes das categorias do eixo X, garantindo melhor legibilidade;
- Remover o retângulo que envolve o gráfico, para uma apresentação mais simples e direta;
- Excluir o eixo Y, pois na próxima atividade vamos trabalhar na exibição dos valores diretamente nas colunas;
- Remover os ticks dos eixos, a fim de limpar o gráfico e focar nos dados essenciais.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'], y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

ax.set_xlabel('') # ocultando rótulo do eixo X
ax.xaxis.set_tick_params(labelsize=12) # aumentando tamanho da fonte das categorias
ax.set_frame_on(False) # removendo retângulo que envolve o gráfico
ax.get_yaxis().set_visible(False) # removendo eixo Y
ax.tick_params(left=False, right=False) # removendo ticks dos eixos

plt.show()
```

#### Exercício 4: Adicionando valores ao gráfico

Sua tarefa será:

Adicionar os valores da média de aluguel de cada região acima das colunas no gráfico. Certifique-se de formatar estes valores em - reais, por exemplo, R$ 3.500, para facilitar a interpretação e proporcionar uma visão clara do custo do aluguel em cada área. Isso - não só enriquecerá a visualização com informações numéricas precisas, mas também permitirá uma comparação imediata entre as - 
diferentes regiões, destacando as variações de preço de forma mais explícita e acessível.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'], y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

ax.set_xlabel('') # ocultando rótulo do eixo X
ax.xaxis.set_tick_params(labelsize=12) # aumentando tamanho da fonte das categorias
ax.set_frame_on(False) # removendo retângulo que envolve o gráfico
ax.get_yaxis().set_visible(False) # removendo eixo Y
ax.tick_params(left=False, right=False) # removendo ticks dos eixos

# adicionando valores ao gráfico
for i, valor in enumerate(df_agrupado['Valor']):
    valor_medio = f'R${valor:,.0f}'.replace(',','.')
    offset = 90
    ax.text(x=i, y= valor + offset, s=valor_medio, fontsize=10, ha='center', va='center')

plt.show()
```

#### Exercício 5: Adicionando textos conclusivos ao gráfico

Sua tarefa será:

Adicionar um texto conclusivo ao lado do gráfico. Esse texto resumirá os dados apresentados e destacará insights relevantes, como a diferença significativa entre os valores de aluguel nas regiões Oeste e Leste. Por exemplo, podemos incluir uma observação sobre como a região Oeste, com o aluguel médio mais alto de R$ 3.590, contrasta com a região Leste, onde o aluguel médio é de apenas R$ 1.697.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'], y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

ax.set_xlabel('') # ocultando rótulo do eixo X
ax.xaxis.set_tick_params(labelsize=12) # aumentando tamanho da fonte das categorias
ax.set_frame_on(False) # removendo retângulo que envolve o gráfico
ax.get_yaxis().set_visible(False) # removendo eixo Y
ax.tick_params(left=False, right=False) # removendo ticks dos eixos

for i, valor in enumerate(df_agrupado['Valor']):
    valor_medio = f'R${valor:,.0f}'.replace(',','.')
    offset = 90
    ax.text(x=i, y= valor + offset, s=valor_medio, fontsize=10, ha='center', va='center')

regiao_maximo = df_agrupado['Regiao'].values[0]
regiao_minimo = df_agrupado['Regiao'].values[-1]

ax.text(x=2.4, y=3300, 
       s=f'O valor do aluguel na região $\\bf{{{regiao_maximo}}}$\n'
         f'é mais que o dobro da região $\\bf{{{regiao_minimo}}}$.'
)

plt.show()
```
