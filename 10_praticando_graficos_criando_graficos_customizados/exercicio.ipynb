#### Exercício 1: Criando o primeiro gráfico

Vamos começar nossa jornada na criação e customização de gráficos com esta atividade, que é apenas o primeiro passo. Aqui, você terá a oportunidade de analisar um conjunto de dados sobre aluguéis de apartamentos em diferentes regiões de São Paulo, explorando como a região influencia o valor médio do aluguel. Essa análise ajudará a identificar padrões ou tendências entre as regiões e os preços de aluguel, fornecendo insights valiosos sobre o mercado imobiliário da cidade.

Sua tarefa será:
- Carregar os dados para análise;
- Agrupar os dados por região e calcular a média dos valores de aluguel;
- Ordenar as regiões em ordem decrescente, do maior para o menor valor médio de aluguel;
- Criar uma figura e definir um eixo, ajustando o tamanho da figura para que os dados fiquem bem visualizados;
- Escolher um gráfico apropriado para comparar os valores médios de aluguel entre as regiões;
- Adicionar um título ao gráfico para torná-lo mais informativo e fácil de interpretar.

Dica: Ao adicionar o título, considere:
- Aumentar o tamanho da fonte para dar destaque;
- Inserir um espaçamento adequado entre o título e o gráfico para melhorar a leitura;
- Posicionar o título no lado esquerdo, pois isso segue o fluxo natural da leitura em idiomas que utilizam o sentido da esquerda para a direita.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'],y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

plt.show()
```

#### Exercício 2: Removendo a saturação do gráfico

Agora é o momento de aprimorar nossa visualização, focando em torná-la mais clara e eficiente. Para isso, vamos trabalhar na remoção da saturação visual, eliminando elementos que não agregam valor ou que podem distrair a atenção.

Sua tarefa será:
- Ocultar o rótulo do eixo X para um visual mais limpo;
- Aumentar o tamanho das fontes das categorias do eixo X, garantindo melhor legibilidade;
- Remover o retângulo que envolve o gráfico, para uma apresentação mais simples e direta;
- Excluir o eixo Y, pois na próxima atividade vamos trabalhar na exibição dos valores diretamente nas colunas;
- Remover os ticks dos eixos, a fim de limpar o gráfico e focar nos dados essenciais.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'], y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

ax.set_xlabel('') # ocultando rótulo do eixo X
ax.xaxis.set_tick_params(labelsize=12) # aumentando tamanho da fonte das categorias
ax.set_frame_on(False) # removendo retângulo que envolve o gráfico
ax.get_yaxis().set_visible(False) # removendo eixo Y
ax.tick_params(left=False, right=False) # removendo ticks dos eixos

plt.show()
```

#### Exercício 3: Adicionando valores ao gráfico

Sua tarefa será:

Adicionar os valores da média de aluguel de cada região acima das colunas no gráfico. Certifique-se de formatar estes valores em - reais, por exemplo, R$ 3.500, para facilitar a interpretação e proporcionar uma visão clara do custo do aluguel em cada área. Isso - não só enriquecerá a visualização com informações numéricas precisas, mas também permitirá uma comparação imediata entre as - 
diferentes regiões, destacando as variações de preço de forma mais explícita e acessível.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(data=df_agrupado, x=df_agrupado['Regiao'], y=df_agrupado['Valor'])

ax.set_title('Valor Médio dos Alugueis por Região', fontsize=14, pad=12, loc='left')

ax.set_xlabel('') # ocultando rótulo do eixo X
ax.xaxis.set_tick_params(labelsize=12) # aumentando tamanho da fonte das categorias
ax.set_frame_on(False) # removendo retângulo que envolve o gráfico
ax.get_yaxis().set_visible(False) # removendo eixo Y
ax.tick_params(left=False, right=False) # removendo ticks dos eixos

# adicionando valores ao gráfico
for i, valor in enumerate(df_agrupado['Valor']):
    valor_medio = f'R${valor:,.0f}'.replace(',','.')
    offset = 90
    ax.text(x=i, y= valor + offset, s=valor_medio, fontsize=10, ha='center', va='center')

plt.show()
```

#### Exercício 5: Focando a atenção onde desejamos

Para finalizar nosso gráfico de forma eficaz, vamos direcionar a atenção do público-alvo para o ponto mais crucial da análise: a notável diferença de valores entre as regiões Oeste e Leste. Podemos estrategicamente utilizar cores para alcançar esse objetivo, destacando essas duas regiões com tons distintos. Assim, ao colorir as barras das regiões Oeste e Leste diferentemente das demais, reforçamos visualmente a disparidade de preços, garantindo que essas informações se destaquem e facilitem a compreensão imediata do impacto da localização nos preços dos aluguéis.

Sua tarefa será:
- Criar uma lista de cores, atribuindo uma cor específica a cada coluna no gráfico. Para enfatizar visualmente a disparidade entre os valores mais altos e mais baixos dos aluguéis, escolha cores específicas para a região com o maior valor (Oeste) e a região com o menor valor (Leste). As demais regiões podem ser representadas em tons mais neutros para manter o foco nas regiões de interesse.
- Ajustar as cores do título e das categorias para tons mais claros. Isso não apenas melhora a legibilidade, mas também adiciona um contraste visual que destaca essas informações importantes sem competir com a informação visual das barras.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv'

df = pd.read_csv(url)

df.head()

df_agrupado = df.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index()

AZUL1, AZUL2, AZUL3, AZUL4, AZUL5 = '#03045e', '#0077b6', "#00b4d8", '#90e0ef', '#CDDBF3'
CINZA1, CINZA2, CINZA3, CINZA4, CINZA5 = '#212529', '#495057', '#adb5bd', '#dee2e6', '#f8f9fa'
VERMELHO1, LARANJA1, AMARELO1, VERDE1, VERDE2, VERDE3 = '#e76f51', '#f4a261',	'#e9c46a', '#4c956c', '#2a9d8f', '#264653'

cores = [
    VERMELHO1 if valor == df_agrupado['Valor'].max() 
    else VERDE3 if valor == df_agrupado['Valor'].min()
    else CINZA1 
    for valor in df_agrupado['Valor']
    ]

fig, ax = plt.subplots(figsize=(8,6))

ax = sn.barplot(
    data=df_agrupado,
    x='Regiao',
    y='Valor',
    hue='Regiao',
    palette=cores
)
ax.set_title('Valor Médio dos Aluguéis por Região', loc='left', fontsize=14, pad=15, weight='bold', color=CINZA2)
ax.set_frame_on(False)
ax.set_xlabel('')
ax.tick_params(length=0)
ax.get_yaxis().set_visible(False)
ax.xaxis.set_tick_params(labelsize=12, labelcolor=CINZA1)

for i, valor in enumerate(df_agrupado['Valor']):
    valor_medio = f'R${valor:,.0f}'.replace(',', '.')
    offset = 100
    ax.text(x=i, y=valor + offset, s=valor_medio, va='center', ha='center', color=CINZA1, fontsize=11)

regiao_maximo = df_agrupado['Regiao'].values[0]
regiao_minimo = df_agrupado['Regiao'].values[-1]

ax.text(x=2.4, y=3300, 
        s=f'O valor do aluguel na região $\\bf{{{regiao_maximo}}}$\n'
          f'é mais que o dobro da região $\\bf{{{regiao_minimo}}}$.',
        color=CINZA2
)

plt.show()
```

#### Exercício 7: Criando um gráfico para analisar uma série temporal

Agora que já carregamos e preparamos os dados, é hora de dar o próximo passo: a criação do nosso gráfico. Vamos iniciar com um design simples e progressivamente aprimorar nossa visualização nas atividades seguintes.

#### Exercício 8: Adicionando anotações ao gráfico

Sua tarefa será:
- Adicionar um texto abaixo do título que especifique o período dos dados analisados, para contextualizar o espectador;
- Adicionar um texto abaixo do gráfico que mencione a fonte dos dados: "Transport for London - TfL, disponível em https://cycling.data.tfl.gov.uk/";
- Elaborar uma conclusão breve sobre os dados, realçando observações como a sazonalidade e os picos de demanda notáveis, associados a eventos específicos;
- Para aprimorar a estética do gráfico e garantir que as anotações adicionadas sejam claramente visíveis sem obstruções, é recomendável remover as bordas desnecessárias que circundam o gráfico. Isso pode ser feito utilizando a função sns.despine() da biblioteca Seaborn.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/contagem_bicicletas.csv'

df = pd.read_csv(url)

df['data_hora'] = pd.to_datetime(df['data_hora'])

df['data'] = df['data_hora'].dt.date

df_agrupado = df.groupby('data')['contagem'].sum().reset_index()

fig, ax = plt.subplots(figsize=(12,5), dpi=100)

ax = sn.lineplot(
        data=df_agrupado,
        x='data',
        y='contagem'
)

ax.set_title('Demanda Diária por Bicicletas em Londres', pad=30, loc='left', fontsize=16, weight='bold')
ax.text(x=0, y=1.04, s=' Registros entre jan/2025 e jan/2027.',transform=ax.transAxes, fontsize=12)
ax.text(x=0, y=-0.14, s=' Fonte dos dados: Transport for London - TfL, disponível em https://cycling.data.tfl.gov.uk/',
        transform=ax.transAxes, fontsize=10)

ax.text(0.36, 0.65,
        'Os picos de demanda em julho e agosto de 2015\n'
        'são atribuídos às greves de metrô. Adicionalmente,\n'
        'observa-se que a demanda por bicicletas aumenta\n'
        'significativamente durante os meses mais quentes\n'
        'ano.\n', 
        transform=ax.transAxes,
        fontsize=12,
        linespacing=1.5)

ax.set_xlabel('')
ax.set_ylabel('Número de bicicletas')
sn.despine()

plt.show()
```

#### Exercício 9: Formatando os valores do eixo Y

Agora, vamos aprimorar ainda mais nossa visualização ajustando os valores no eixo Y para melhorar a clareza e a leitura dos dados.

Sua tarefa será:
- Formatar os valores numéricos, utilizando o formato abreviado (por exemplo, '10K' para 10.000). Isso torna o gráfico menos congestionado e mais fácil de analisar, especialmente quando estamos lidando com grandes quantidades.
- Ajustar o início do eixo Y para 0, pois essa é uma prática padrão em muitas visualizações de dados, proporcionando uma perspectiva mais realista e compreensível das variações e das diferenças entre os dados apresentados.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/contagem_bicicletas.csv'

df = pd.read_csv(url)

df['data_hora'] = pd.to_datetime(df['data_hora'])

df['data'] = df['data_hora'].dt.date

df_agrupado = df.groupby('data')['contagem'].sum().reset_index()

fig, ax = plt.subplots(figsize=(12,5), dpi=100)

ax = sn.lineplot(
        data=df_agrupado,
        x='data',
        y='contagem'
)

ax.set_title('Demanda Diária por Bicicletas em Londres', pad=30, loc='left', fontsize=16, weight='bold')
ax.text(x=0, y=1.04, s=' Registros entre jan/2025 e jan/2027.',transform=ax.transAxes, fontsize=12)
ax.text(x=0, y=-0.14, s=' Fonte dos dados: Transport for London - TfL, disponível em https://cycling.data.tfl.gov.uk/',
        transform=ax.transAxes, fontsize=10)

ax.text(0.36, 0.65,
        'Os picos de demanda em julho e agosto de 2015\n'
        'são atribuídos às greves de metrô. Adicionalmente,\n'
        'observa-se que a demanda por bicicletas aumenta\n'
        'significativamente durante os meses mais quentes\n'
        'ano.\n', 
        transform=ax.transAxes,
        fontsize=12,
        linespacing=1.5)

plt.ylim((0, 70000))
formatter = FuncFormatter(lambda val, pos: f'{int(val/1000)}K')
plt.gca().yaxis.set_major_formatter(formatter)
ax.set_xlabel('')
ax.set_ylabel('Número de bicicletas')
sn.despine()

plt.show()
```
